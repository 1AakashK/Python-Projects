```python
def add_labels(data):
  max_time_cycles=[]
  labels=[]
  for i in data['unit-number']:
    max_time_cycles.append(((data[data['unit-number'] == i]['time_in_cycles']).values)[-1])
  data['max_time_cycles'] = max_time_cycles
  data['life_ratio'] = data['time_in_cycles'].div(data['max_time_cycles'])
  for i in range (0,len(data)):
    if np.array(data['life_ratio'])[i] <= 0.6:
      labels.append(0)
    else :
      labels.append(1)
  data = data.drop(columns=['max_time_cycles', 'life_ratio'])
  data['labels'] = labels
  return data
```

```python
x = RUL.drop(['labels'], axis=1)
x.head()
```

```python
y = RUL['labels']
y
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.20, random_state=42)
```

###SVM

```python
clf = svm.SVC(kernel='linear')
```

```python
clf.fit(x_train, y_train)

```

```python
# Predict the class labels of test data
y_pred = clf.predict(x_test)

# Print the accuracy of the classifier
accuracy = clf.score(x_test, y_test)
print("Accuracy:", accuracy)
```

```python
print(metrics.classification_report(y_test, y_pred))
```

###Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier()

rfc.fit(x_train, y_train)
```

```python
y_pred = rfc.predict(x_test)
```

```python
print(metrics.accuracy_score(y_test, y_pred))
```

```python
print(metrics.classification_report(y_test, y_pred))
```

###Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()

lr.fit(x_train, y_train)
```

```python
y_pred = lr.predict(x_test)
```

```python
print(metrics.accuracy_score(y_test, y_pred))
```

```python
print(metrics.classification_report(y_test, y_pred))
```

###Decision Tree
